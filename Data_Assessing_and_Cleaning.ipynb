# DATA CLEANING





### Data Quality Dimensions

- Completeness -> is data missing?
- Validity -> is data invalid -> negative height -> duplicate patient id
- Accuracy -> data is valid but not accurate -> weight -> 1kg
- Consistency -> both valid and accurate but written differently -> New Youk and NY

 
-` NOTE :-` once you get all the problems in data labled them by above lables. `LABELING done in only dirty data`.


#### Order Of Severity :--
- WHICH DATA IS MORE DENGEROUS
    - Completeness <- Validity <- Accuracy <- Consistency

- WHY WE `LABLED` THE DIRTY DATA ? 
    - because There is order of Data cleaning which has to be followed in data cleaning process. 

#### Data cleaning order :--

- Quality - Completeness`
- Tideness - Messy data
- Quality - Validity
- Quality - Accuracy
- Quality - Consistency

#### Steps Involved in Data cleaning 

- Define --> define the problem's Solution . `in a line/comment` - the way you are going to solve the issue
- code -->  write the code of that solution
- Test --> test the code on you data

`Always make sure you have a copy of your data testing the sloution on copy data. `

### Using For Loop to 

```python
temp_treatment_df= treatments_df.copy()
temp_treatment_df
```

```python
typee=[]
for name in temp_treatment_df['auralin']:
    if name=='-':
        typee.append('novodra')
    else:
        typee.append('auralin')
typee
```

```python
temp_treatment_df['type']=''
temp_treatment_df
```

```python
for i in range(temp_treatment_df.shape[0]):
    temp_treatment_df['type'][i]=typee[i]
temp_treatment_df
```

```python
## Replacing from each row of a particular column : it doesn't replace element from value
temp_treatment_df['auralin'].replace('-',np.nan,inplace=True)
temp_treatment_df['novodra'].replace('-',np.nan,inplace=True)
```

```python
temp_treatment_df['doses']=temp_treatment_df['auralin'].combine_first(temp_treatment_df['novodra'])
```

```python
temp_treatment_df.drop(['auralin','novodra','doses'],axis=1,inplace=True)
```

```python
temp_treatment_df['start_dose']=temp_treatment_df['doses'].str.split('-').str.get(0)
temp_treatment_df['end_dose']=temp_treatment_df['doses'].str.split('-').str.get(1)
```

```python
## Replacing elments from the value of each row
temp_treatment_df['start_dose']=temp_treatment_df['start_dose'].str.replace('u','')
```

```python
temp_treatment_df['end_dose']=temp_treatment_df['end_dose'].str.replace('u','')
```

```python
temp_treatment_df['start_dose']=temp_treatment_df['start_dose'].astype(int)
temp_treatment_df['end_dose']=temp_treatment_df['end_dose'].astype(int)
```

## Using melt() method

```python
## melting the columns in which i want to put the column in a data frame
treatments_df=treatments_df.melt(id_vars=['given_name', 'surname', 'hba1c_start',	'hba1c_end', 'hba1c_change'], var_name='type', value_name='doses')
```

```python
treatments_df=treatments_df[treatments_df['doses']!= '-']
```

```python
## splitting the doses values by '-' and put the splitteed values in two different column called start_dose and end_dose
treatments_df['start_dose']=treatments_df['doses'].str.split('-').str.get(0)
treatments_df['end_dose']=treatments_df['doses'].str.split('-').str.get(1)
treatments_df
```

```python
## replacing the 'u' in start & end dose columnvalues
treatments_df['start_dose']=treatments_df['start_dose'].str.replace('u','')
treatments_df['end_dose']=treatments_df['end_dose'].str.replace('u','')
treatments_df
```

```python
## changing data type from object to int of start dose and end dose column
treatments_df['start_dose']=treatments_df['start_dose'].astype(int)
treatments_df['end_dose']=treatments_df['end_dose'].astype(int)
treatments_df.info()
```
